```python
#CARREGANDO AS BIBLIOTECAS
import pandas as pd
```

```python
#CARREGANDO OS DADOS EM CSV
arquivo = pd.read_csv('csv/datasets.csv')
```

```python
arquivo.head()
```

```python
#ARMAZENANDO DENTRO DE UMA VARIAVEL, E DEPOIS SUBSTITUI O RED POR 0
arquivo['style'] = arquivo['style'].replace('red', 0)
```

```python
#ARMAZENANDO DENTRO DE UMA VARIAVEL, E DEPOIS SUBSTITUI O WHITE POR 1
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
arquivo
```

```python
arquivo.shape
```

```python
#SEPARADO AS VARIAVEIS ENTRE PREDITORIAS E VARIAVEL ALVO
y = arquivo['style']
x = arquivo.drop('style', axis = 1)
```

```python
x.shape
```

```python
y.shape
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

#CRIANDO OS CONJUNTOS DE DADOS DE TREINOS E TESTE:
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
x_teste.shape
```

```python
y_teste.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#CRIAÇÃO DO MODELO:
modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino, y_treino)

#IMPRIMMINDO RESULTADOS:
resultado = modelo.score(x_teste, y_teste)
print('Acurácia: ', resultado)
```
